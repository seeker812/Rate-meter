# **Feature Selection**

```python
corr_matrix = df.corr().abs()['Price']
high_corr_features = corr_matrix[corr_matrix > 0.5].index.tolist()
high_corr_features.remove('Price')  

X = df[high_corr_features]
Y = df['Price']
```

# **Normalization**

> There are several advantages in normalizing features, for example it facilitates the convergence of the cost function towards its minimum.
StandardScaler: Standardizes each variable X: the mean is zero and the standard deviation is equal to 1.
These methods are very sensitive to the outliers

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)
```

## **Splitting the dataset into training and testing sets**

```python
import numpy as np
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_normalized, Y, test_size=0.2, random_state=42)

y_train = np.array(y_train).reshape(-1, 1)
y_test = np.array(y_test).reshape(-1, 1)
```

## **Initialize parameters - weights and biases**

> Here we are creating two  layers one hidden layer and one output layer 
**
* weights and bias for one hideen layer is weight (6,8) here 6 are i/p for hidden layer and 8 are neurons & bias (1,8) here we have 8 biases for each neurons
* weights and bias for one output layer is weight (8,1) here 8 are i/p for hidden layer and 1 are neurons & bias (1,1) here we have 1 bias for each nbeuron
***

```python
n_inputs = X_train.shape[1]  # Number of features
n_hidden = 8  # Number of nodes in the hidden layer
n_outputs = 1 

 
W1 = np.random.randn(n_inputs, n_hidden) * 0.01
W2 = np.random.randn(n_hidden, n_outputs) * 0.01
b1 = np.zeros((1, n_hidden))
b2 = np.zeros((1, n_outputs))
```

## **Model making**

```python
# Activation function
def relu(x):
    return np.maximum(0, x)

# Mean Squared Error
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Forward propagation
def forward_prop(X):
    Z1 = np.dot(X, W1) + b1
    A1 = relu(Z1)
    Z2 = np.dot(A1, W2) + b2
    return Z2, A1

# Backpropagation
def back_prop(X, y, output, A1):
    m = X.shape[0]
    dZ2 = output - y
    dW2 = np.dot(A1.T, dZ2) / m
    db2 = np.sum(dZ2, axis=0, keepdims=True) / m
    dZ1 = np.dot(dZ2, W2.T) *(A1 > 0)
    dW1 = np.dot(X.T, dZ1) / m
    db1 = np.sum(dZ1, axis=0, keepdims=True) / m
    return dW1, db1, dW2, db2

# Gradient Clipping 
def clip_gradients(dW1, db1, dW2, db2, clip_value=1):
    dW1 = np.clip(dW1, -clip_value, clip_value)
    db1 = np.clip(db1, -clip_value, clip_value)
    dW2 = np.clip(dW2, -clip_value, clip_value)
    db2 = np.clip(db2, -clip_value, clip_value)
    return dW1, db1, dW2, db2
```

# **Training the model**

```python
learning_rate = 0.8  
n_epochs = 900

for epoch in range(n_epochs):
    output, A1= forward_prop(X_train)
    dW1, db1, dW2, db2 = back_prop(X_train, y_train, output,A1)

#     Gradient Clipping
    dW1, db1, dW2, db2 = clip_gradients(dW1, db1, dW2, db2)

    # Update weights and biases
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2

    # Print cost every 100 iterations
    if epoch % 100 == 0:
        cost = mse(y_train, output)
        print("Epoch",epoch,"  Loss:" ,cost)
```

# **Prediction**

```python
predictions, _ = forward_prop(X_test)
print(predictions)
print(y_test)
```

# **Evaluation**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score

print("MAE: ",mean_absolute_error(y_test, predictions))
print("MSE: ",mean_squared_error(y_test, predictions))
print("R2: ",r2_score(y_test, predictions))
```
